```python

from confluent_kafka import Producer, Consumer
import pandas as pd
import time
```

```python
x_for_pred1 = pd.read_csv("store1_data.csv")

x_for_pred1 = x_for_pred1.iloc[::-1]

P_1 = Producer({'bootstrap.servers':' localhost:9092'})

for i, j in x_for_pred1.iterrows():
        a = str(j.to_list()).split('[')[1].split(']')[0]
        a = a.split(',')
        a = [int(float(b)) for b in a] 
        P_1.produce(topic = '1' , key ='' , value = "Sample," + str(a).split('[')[1].split(']')[0])
        print("new sample for store1")
        #print(str(j.to_list()).split('[')[1].split(']')[0])
        time.sleep(200)#10 min rest
 
```
