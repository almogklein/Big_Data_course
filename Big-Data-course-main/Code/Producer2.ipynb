```python
from confluent_kafka import Producer, Consumer
import pandas as pd
import time
```

```python
x_for_pred2 = pd.read_csv("store2_data.csv")

x_for_pred2 = x_for_pred2.iloc[::-1]

P_2 = Producer({'bootstrap.servers':' localhost:9092'})

for i, j in x_for_pred2.iterrows():
        a = str(j.to_list()).split('[')[1].split(']')[0]
        a = a.split(',')
        a = [int(float(b)) for b in a] 
        P_2.produce(topic = '2' , key ='' , value = "Sample," + str(a).split('[')[1].split(']')[0])
        print("new sample for store1")
        #print(str(j.to_list()).split('[')[1].split(']')[0])
        time.sleep(200)#10 min rest
```
